```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y):
    return x + 2*y

def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
```

```python
N = 20
t = np.linspace(0,10,N)

h = t[1] - t[0]

##yext = Solucion(t)
#plt.scatter(t,yext)
```

```python
def GetEuler(x,y,h):

    for i in range(1,len(x)):

        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1

y1 = np.zeros_like(t)
y1[0] = 0

GetEuler(t,y1,h)
#plt.plot(t,yext,"o")
#plt.plot(t,y1,"o")
```

```python
def GetEuler2(x,y,h):

    for i in range(1,len(x)):

        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)

        y[i] = y[i-1] + 0.5*h*(k1+k2)


y2 = np.zeros_like(t)
y2[0] = 0.
GetEuler2(t,y2,h)
#plt.plot(t,yext,"o",label='Euler_exacta')
##plt.plot(t,y1,"o",label='Euler1')
#plt.plot(t,y2,"o",label='Euler2')
#plt.legend()
```

```python
def GetEuler3(x,y,h):

    for i in range(1,len(x)):

        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
        k3 = f(x[i-1]+h,y[i-1] - h*k1 + 2*h*k2)

        y[i] = y[i-1] + (h/6)*(k1+4*k2+k3)


y3 = np.zeros_like(t)
y3[0] = 0.
GetEuler3(t,y3,h)
##plt.plot(t,yext,"o",label='Euler_exacta')
#plt.plot(t,y1,"o",label='Euler1')
#plt.plot(t,y2,"o",label='Euler2')
#plt.plot(t,y3,"o",label='Euler3')
#plt.legend()
```
