```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = -1
u_0 = 1

largo = 15
t_1 = np.arange(0,largo,1.1)
t_2 = np.arange(0,largo,1.5)
t_3 = np.arange(0,largo,1.9)

print(t_3)
```

```python
def f(u,alpha):
  return alpha*u


```

```python

def get_euler(f,u_0,t):
  h = t[1] - t[0]

  u = np.zeros_like(t)

  u[0] = u_0
  #k1 = 0

  for i in range(1,len(t)):

    u[i] = ((1 + (alpha * h))**i) * u_0

  return u
```

```python
U_1 = get_euler(f,u_0,t_1)
U_2 = get_euler(f,u_0,t_2)
U_3 = get_euler(f,u_0,t_3)
#print(len(U_1))
#print(len(t_1))
plt.plot(t_1,U_1,label='Paso 1.1')
plt.plot(t_2,U_2,label='Paso 1.5')
plt.plot(t_3,U_3,label='Paso 1.9')

t = np.arange(0,largo,0.1)
plt.plot(t,np.exp(-t),label='solucion exacta')
plt.xlabel('t')
plt.ylabel('u(t)')

plt.legend()
plt.show
```

```python
print("Es apreciable que la solucion que nos da nuestro metodo iterativo es oscilante")
print("con pasos temporales de 1.1 s, 1.5 s, 1.9 s los cuales son claramente mayores a 1s.")
print("Ademas, son mas oscilantes en tanto el paso sea mayor, como se ve en la grafica.")
```
